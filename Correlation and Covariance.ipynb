```python
corr_msrp = msrp.corr(numeric_only = True)
```

```python
sns.heatmap(corr_msrp, annot = True, cmap= 'coolwarm')
```

```python
spear_corr = msrp.corr(method="spearman", numeric_only = True)
```

```python
sns.heatmap(spear_corr, annot = True, cmap= 'coolwarm')
```

```python
sns.scatterplot(data = msrp, x= 'Engine HP', y= 'MSRP')

sns.regplot(data = msrp, x= 'Engine HP', y= 'MSRP', scatter = False, color = 'red')
```
