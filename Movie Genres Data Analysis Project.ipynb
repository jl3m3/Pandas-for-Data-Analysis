## Research Questions (Q):

### 1. Which genres are the most common?

```python
movies_common = pd.DataFrame(movies_genre.groupby('genres_split').original_title.nunique()).sort_values('original_title', ascending = False)
movies_common
```

# Research Hypotheses (H)

1. The best movies according to vote avg. return high profit and revenue.

2. The best movies according to popularity return high profit and revenue.

3. Highly budgeted movies return high revenue and profit.

4. Highly budgeted movies have a high popularity.


```python
# 1.The best movies according to vote avg. return high profit and revenue.

movies.drop_duplicates(inplace=True)
movies['profit'] = movies ['revenue'] - movies ['budget']
movies_rese = movies[['popularity','budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_date', 'vote_average', 'vote_count', 'profit']]
movies_rese.head()
```

```python
#Checking correlation, needed to add numeric_only otherwise error

movies_counted = movies_rese[movies_rese['vote_count'] >= 50]

movies_counted.corr(method = 'spearman', numeric_only = True)
```

```python
import seaborn as sns

sns.regplot(x = 'vote_average',  y = 'revenue', data = movies_counted, line_kws = {"color": "red"})
```

```python
# 2. The best movies according to popularity return high profit and revenue.

movies_counted.corr(method = 'spearman', numeric_only = True)
```

```python
import matplotlib.pyplot as plt
sns.regplot(x = 'popularity',  y = 'revenue', data = movies_counted, line_kws = {"color": "red"})
plt.figure(figsize =(20,5))
plt.show()
```

```python
# 3. Highly budgeted movies return high profit.
# 4. Highly budgeted movies have a high popularity.

movies_rese.head()
```

```python
sns.regplot(x = 'budget',  y = 'profit', data = movies_counted, line_kws = {"color": "red"})
plt.figure(figsize =(20,5))
plt.show()
```

```python
# 5. Look at Profit per Genre per year
# Lets change date to release year and do a little time serie analysis.

movies.drop_duplicates(inplace=True)
movies['profit'] = movies ['revenue'] - movies ['budget']
movies_year = movies[['popularity','budget', 'revenue', 'original_title', 'runtime', 'genres', 'release_year', 'vote_average', 'vote_count', 'profit']]
movies_year.head()
```

```python
from pandas import Series, DataFrame

movies_year['genres'].str.split('|').apply(Series, 1). stack()

# Adding genres to series and stack those under individual movie.
```

```python
split = movies_year['genres'].str.split('|').apply(Series, 1).stack()
split.index = split.index.droplevel(-1)
split.name = 'genres_split'
del movies_year['genres']
movies_year = movies_year.join(split)

# Dropping the additional index
# Delete old genres column
# Adding new genres_split to data movies_genre
```

```python
movies_year.head()
```

```python
time_genre = pd.DataFrame(movies_year.groupby(['release_year', 'genres_split'])['profit'].mean(numeric_only = True))
```

```python
#Pivoting profit to years and genres

final_countdown = pd.pivot_table(time_genre, values = 'profit', index = ['genres_split'], columns = ['release_year'])
```

```python
sns.set(rc = {'figure.figsize': (15,10)})
sns.heatmap(final_countdown, cmap = 'YlGnBu', linewidths = .5)
plt.title ('Genres by Profit per Year')

#Visualizing data to heatmap
#Cmap styles 'coolwarm' or 'YlGnBu'
```
