```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Housing data pd.read_csv

housing = pd.read_csv(r"C:\Users\jl3m3\repos\Data\real_estate.csv", encoding = "ISO-8859-1")

housing
```

```python
housing_filtered = housing[housing['state'].isin (['Massachusetts', 'Puerto Rico', 'New York'])]
```

```python
#Using Seaborn

sns.barplot(housing_filtered, x = 'state' ,y='price', estimator = 'median')
plt.plot()
```

```python
#Using Pandas

housing_agg = housing_filtered.groupby('state')['price'].mean().reset_index()
```

```python
#Using Pandas

housing_agg.plot(x='state', y = 'price', kind = 'bar')
```

```python
#TEsting Seaborn styles "dark", "darkgrid" and so on
sns.set_style('dark')
```

```python
sns.set_palette('deep')

housing_filt = housing[housing['house_size'] < 10000]

sns.regplot(data = housing_filt, x = 'house_size', y = 'price', line_kws ={'color' : 'red'})

plt.ticklabel_format(style='plain', axis = 'y')

plt.show()
```
