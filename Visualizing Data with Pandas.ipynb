```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\janne.lemettinen\repos\Data\all_stocks_5yr.csv" , encoding = "ISO-8859-1")
df
```

```python
#Lineplot
df.plot(x = 'date', y = 'close', ylabel = 'Closing Price', xlabel = 'Date of Price', title = 'Stock Closing Prices')
```

```python
#Company is American Airlines 'AAL'
company_aal = df[df['Name'] == 'AAL']

company_aal.plot(x = 'date', y = 'close', ylabel = 'Closing Price', xlabel = 'Date of Price', title = 'Stock Closing Prices')
```

```python
df
```

```python
#Barplot

df.plot(x='Name', y = 'Volume', kind = 'bar')
```

```python
# Filrering results to 3 companies AAL, AAP, AAPL

agg_df = df.groupby('Name')['volume'].sum().reset_index()

agg_df = agg_df[agg_df['Name'].isin(['AAL', 'AAP', 'AAPL'])]
agg_df
```

```python
#Barplot

agg_df.plot(x='Name', y = 'volume', kind = 'bar')
```

```python
#Scatterplot using company_aal filtered data. Open and close prices.

company_aal.plot(x = 'open', y = 'close', kind = 'scatter')
```

```python
#Histogram using company_aal filtered data. Open and close prices.

company_aal['high'].plot.hist(bins = 25)
```
