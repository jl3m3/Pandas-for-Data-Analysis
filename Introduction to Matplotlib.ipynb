```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker ='^', markersize = 10, linestyle = '--', markeredgecolor = 'red')

plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```

```python
x = [1,2,3]
y = [5,10,20]

plt.plot(x,y, label = 'Showers', color = 'green', marker ='^', markersize = 10, linestyle = '--', markeredgecolor = 'red')

z = [1,5,10]
plt.plot(x,z, label = 'Baths', color = 'red', marker ='.', markersize = 15, linestyle = '--', markeredgecolor = 'green')

plt.title('Matplotlib Viz', fontdict = {'fontname': 'Times New Roman', 'fontsize': 30})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

plt.xticks([1,2,3])
plt.yticks([0,5,10,15,20])

plt.legend()

plt.show()
```
